```python
keep=[1,2,4,6,8,10,12,14,16]  # números de columnas que deseas mantener
dontKeep=set(hdi.columns.to_list())-set(hdi.columns[keep])
hdiCleaned=hdi.drop(columns=dontKeep)

hdiCleaned
```

```python
hdihdi=hdiCleaned[9:203]
hdihdi
```

```python
change1={'Unnamed: 1':"HDI Rank"}
hdihdi.rename(columns=change1,inplace=True)
hdihdi
change2={'Unnamed: 2':"Country"}
hdihdi.rename(columns=change2,inplace=True)
hdihdi
change3={'Unnamed: 4':"Human Development Index (HDI) 2021"}
hdihdi.rename(columns=change3,inplace=True)
hdihdi
change4={'Unnamed: 6':"Life expectancy at birth 2021"}
hdihdi.rename(columns=change4,inplace=True)
hdihdi
change5={'Unnamed: 8':"Expected years of schooling - 2021"}
hdihdi.rename(columns=change5,inplace=True)
hdihdi
change6={'Unnamed: 10':"Mean years of schooling - 2021"}
hdihdi.rename(columns=change6,inplace=True)
hdihdi
change7={'Unnamed: 12':"Gross national income (GNI) per capita - 2021"}
hdihdi.rename(columns=change7,inplace=True)
hdihdi
change8={'Unnamed: 14':"GNI per capita rank minus HDI rank - 2021"}
hdihdi.rename(columns=change8,inplace=True)
hdihdi
change9={'Unnamed: 16':"HDI RANK - 2020"}
hdihdi.rename(columns=change9,inplace=True)
hdihdi
```

```python
objectCols=hdihdi.select_dtypes(include=['object']).columns
objectCols
```

```python
hdihdi.loc[:,objectCols]=hdihdi.loc[:,objectCols].apply(lambda col:col.str.strip())
hdihdi
```
